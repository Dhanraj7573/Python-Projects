```python
import pandas as pd
from IPython.display import display
#get the dataset
flights_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv"
gapminder_url = "https://raw.githubusercontent.com/kirenz/datasets/master/gapminder.csv"

flights = pd.read_csv(flights_url)
gap = pd.read_csv(gapminder_url)

display(flights.head())
display(gap.head())
```

```python
#Check if any values are missing:
display(flights.isnull().sum())
display(gap.isnull().sum())
#number of unique years
display(flights.year.unique())
display(gap.year.unique())
#find the comman years between two dataset
a = set(flights.year.unique())
b = set(gap.year.unique())
c = a.intersection(b)
print(c)
#filter the years

display(flights[flights.year.isin(c)])
display(gap[gap.year.isin(c)])
display(gap[gap.year == 1952])
display(flights[(flights.month == 'January') & (flights.passengers >= 350)])
```

```python
display(flights.sort_values('passengers',ascending=False))
display(flights.loc[flights['year'] == 1950].sort_values('passengers', ascending=True))
```

```python
# compute the passenger per year
display(flights.groupby('year').agg({'passengers':['sum','mean','min','max']}))

# which year, month has the highest passanger
display(flights.groupby(['year','month']).agg({'passengers':['max']}))
```

```python
display(gap.head())
```

```python
flights_pivot = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc=['sum','mean'])
display(flights_pivot.head())
```

```python
gap_pivot = gap.pivot_table(index='year', columns='continent', values='pop', aggfunc='mean')
display(gap_pivot.head())
```

```python
flights_pivot_multi_agg = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc=['mean', 'min', 'max'])
display(flights_pivot_multi_agg.head())
```

```python
flights_melted = flights.melt(id_vars=['year', 'month'], value_vars='passengers', var_name='metric', value_name='passenger_count')
display(flights_melted.head())

a = flights.melt(id_vars=['year','month'], value_vars='passengers', var_name='metric', value_name='passenger_count')
display(a.head())
```

```python
flights_complex_pivot = flights.pivot_table(
    index='year',
    columns='month',
    values='passengers',
    aggfunc=['sum', 'mean', 'min', 'max'],
    margins=True,
    margins_name='All' # Name for the totals row/column
)
display(flights_complex_pivot.head())
```

```python
gap_complex_pivot = gap.pivot_table(
    index='year',
    columns='continent',
    values=['lifeExp', 'pop', 'gdpPercap'],
    aggfunc={'lifeExp': 'mean', 'pop': 'sum', 'gdpPercap': 'mean'},
    margins=True,
    margins_name='All'
)
display(gap_complex_pivot.head())
```

```python
gap_melted = gap.melt(id_vars=['country', 'continent', 'year'], value_vars=['lifeExp', 'pop', 'gdpPercap'], var_name='metric', value_name='value')
display(gap_melted.head())
display(gap.head())
```
